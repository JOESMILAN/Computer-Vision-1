```python
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = 'C:/Users/Pavan.M/PCL1/Tablice.v5i.yolov5pytorch/yolov5/runs/detect/exp3/0_18_hr_png_jpg.rf.212f08e28933a13fefe99def505142b9.jpg'
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(image_path)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(image_path)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left, font, 0.5,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread(image_path)
spacer = 100
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),4)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),4,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img)
plt.show()
```
